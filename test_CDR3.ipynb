```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
r=np.array([289,276,270,240,234,233,231,228,225,219])
#r=np.array([424467,7820,3199,3168,2605,2586,2044,1863,1574,1537])
r=r/sum(r)
n=len(r)
```

```python
p=r
entropy=-sum(p*np.log2(p))
print('entropy={:.3g}'.format(entropy))
```

```python
k=np.arange(n+1)
k
```

```python
x=k/n
x
```

```python
y = np.zeros(n+1)
y[1]=r[0]
for i in range(2,n+1):
    y[i]=y[i-1]+r[i-1]
```

```python
y
```

```python
y0 = np.zeros(n+1)
y0=k/n*sum(r)
y0
```

```python
plt.figure(figsize=(6,6))
plt.plot(x*100,y/max(y)*100,'b.-')
plt.plot(x*100,y0/max(y0)*100,'g.--')
plt.grid(True)
plt.rc('grid', linestyle="--", color='black')
plt.xlabel("Percentage of total reads")
plt.ylabel("Percentage of unique CDR3s")
#plt.title('D.I.={:.2g}'.format(DI))
```
